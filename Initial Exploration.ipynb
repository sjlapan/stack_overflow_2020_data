```python
dev_list = employed_df['DevType'].tolist()
dev_list[0]
```

```python
employed_df.dropna(subset=['DevType'], axis = 0, inplace=True)
```

```python
ds_string = 'Data scientist or machine learning specialist'
da_string = 'Data or business analyst'
de_string = 'Engineer, data'
```

```python
employed_df.loc[
    employed_df["DevType"].str.contains(ds_string) == True,
    'DataScientist'
] = 1
employed_df['DataScientist'].fillna(value=0, axis=0, inplace=True)

employed_df.loc[
    employed_df["DevType"].str.contains(da_string) == True,
    'DataAnalyst'
] = 1
employed_df['DataAnalyst'].fillna(value=0, axis=0, inplace=True)

employed_df.loc[
    employed_df["DevType"].str.contains(de_string) == True,
    'DataEngineer'
] = 1
employed_df['DataEngineer'].fillna(value=0, axis=0, inplace=True)

employed_df.head()
```

```python
# Now we can drop 'DevType'

employed_df.drop(columns=['DevType'], inplace=True)
```
