```python
def unique_val_extractor(df_col):
    '''
    INPUT
    df_col - a column from a pandas dataframe
        that contains strings of values separated by ';'
        
    OUTPUT
    unique_list - a single list of unique substring values from
        the column
        
    This function uses the following steps:
    1. Initialize empty list unique_list
    2. remove null values from df_col
    Then, for each row/item in the column:
    3. Split string on the ';' and add each substring to a list
    4. Check if each substring in the list is in unique_list
    5. If it is not, append the substring to unique_list
    '''
    unique_list = []
    # Remove NaN/null values from column
    df_col = df_col[df_col.notnull()]
    
    # Iterate over pandas column
    for row in df_col:
        # Split substrings into list items
        row_list = row.split(';')
        # Iterate over list of substrings to compare to unique_list
        for i in row_list:
            if (i in unique_list):
                print('Duplicate')
            else:
                unique_list.append(i)
                print('Appended')
                print(len(unique_list))
    return unique_list
```

```python
def skill_counter(df, col, new_col_name):
    '''
    INPUT
    df - a pandas dataframe
    col - a column in the dataframe that contains a string of
        values separated by ';'
    new_col_name - a string value to use as the name of a new column
        
    OUTPUT
    df - The original dataframe is returned with an additional column
    
    1. Remove null rows with null values in the target column.
    2. Create an empty new column
    3. Loop through the rows, splitting the string in the target column
       on the ';'
    4. Assign the length of the resulting list as a value to the new column
    '''
    # Remove NaN/null values from column
    df = df[df[col].notnull()]
    
     # Initialize new column
    df[new_col_name] = ''
    
    for index, row in df.iterrows():
        # Split substrings into list items
        row_list = row[col].split(';')
        # Add length of list as value to new column
        df.loc[index, new_col_name] = len(row_list)
    return df
```

```python
# Write a function that will run a linear model on any X, y combo

def linear_model(X, y):
    '''
    INPUT
    X - Predictor variables
    y - Response variable
    
    OUTPUT
    X_train - X training set
    X_test - X test set
    y_train - y training set
    y_test - y test set
    lm_model - The linear model
    r2_test - The R2 statistic for the model fit

    '''
    X_train, X_test, y_train, y_test = train_test_split(
                                                X, 
                                                y, 
                                                test_size=0.30,
                                                random_state=43
                                        )
    lm_model = LinearRegression(normalize=True)
    try:
        lm_model.fit(X_train, y_train)
    except Exception as e:
        print(e)
    
    y_test_preds = lm_model.predict(X_test)
    r2_test = r2_score(y_test, y_test_preds)
    
    return X_train, X_test, y_train, y_test, lm_model, r2_test
```

```python
# Re-write the function to remove the text values from YearsCode

def clean_data(df):
    '''
    INPUT
    df - pandas dataframe
    
    OUTPUT
    X - matrix of predictor variables
    y - response vector
    
    1. Drop all the rows with no ConvertedComp
    2. Drop rows with non integer values for YearsCode
    3. Create X as all of the columns you want as predictors
    4. Create y as the ConvertedComp column
    5. Fill in NAs with column means for X columns
    '''
    
    df = df.dropna(subset=['ConvertedComp'], axis=0)
    
    df = df[df['YearsCode'].apply(lambda x: x.isnumeric())]
    
    X = df[[
        'YearsCode',
        'DataScientist',
        'DataEngineer',
        'DataAnalyst',
        'LanguageCount',
        'DatabaseCount'
    ]]
    
    y = df['ConvertedComp']
    
    X = X.fillna(X.mean())
    
    return X, y
```
